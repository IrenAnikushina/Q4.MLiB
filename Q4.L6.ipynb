```python
data.label = data.label.map({'female': 0, 'male': 1})
data["label"].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler

# для проведения стандартизации разобъем выборку на трэйн и тест

from sklearn.model_selection import train_test_split

x_data = data.iloc[:,:-1]
y_data = data.iloc[:,-1]

scaler = StandardScaler()

x_data = scaler.fit_transform(x_data)
```

##### 3) Обучить любой классификатор (какой вам нравится)

```python
# Разбиваем выборку на тренировочную и тестовую части и обучаем модель

x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=7)
```

```python
# обучим градиентный бустинг

from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier()

model.fit(x_train, y_train)
y_predict = model.predict(x_test)
```

```python
# словарь для записывания результатов
models_results = {
    'model': [],
    'f1': [],
    'recall': [],
    'precision': []
}
```

```python
from sklearn.metrics import recall_score, precision_score, roc_auc_score, accuracy_score, f1_score

def evaluate_results(y_test, y_predict):
    print('Classification results:')
    f1 = f1_score(y_test, y_predict)
    print("f1: %.2f%%" % (f1 * 100.0)) 
    recall = recall_score(y_test, y_predict, average='binary')
    print("recall: %.2f%%" % (recall * 100.0)) 
    precision = precision_score(y_test, y_predict, average='binary')
    print("precision: %.2f%%" % (precision * 100.0)) 
    return f1, recall, precision

    
f1, recall, precision = evaluate_results(y_test, y_predict)
```

```python
models_results['model'].append('GB')
models_results['f1'].append(f1)
models_results['recall'].append(recall)
models_results['precision'].append(precision)
```

##### 4) Разделить ваш набор данных на два множества: P (positives) и U (unlabeled). Причем брать нужно не все положительные (класс 1) примеры, а только лишь часть

```python
# "прячем" часть данных

mod_data = data.copy()
#get the indices of the positives samples
pos_ind = np.where(mod_data.iloc[:,-1].values == 1)[0]
#shuffle them
np.random.shuffle(pos_ind)
# leave just 25% of the positives marked
pos_sample_len = int(np.ceil(0.9 * len(pos_ind)))
print(f'Using {pos_sample_len}/{len(pos_ind)} as positives and unlabeling the rest')
pos_sample = pos_ind[:pos_sample_len]
```

##### 5) Применить random negative sampling для построения классификатора в новых условиях

```python
# Создаем столбец для новой целевой переменной, где у нас два класса - P (1) и U (-1)
# получаем 396 позитива и 2772 без разметки --> -1.

mod_data['class_test'] = -1
mod_data.loc[pos_sample,'class_test'] = 1
print('target variable:\n', mod_data.iloc[:,-1].value_counts())
```

```python
mod_data.tail(10)
```

```python
#  Отделими [:-2] как истиный класс для проверки, и [:-1] как данные для входной разметки PUL

x_data = mod_data.iloc[:,:-2].values # только X 
y_labeled = mod_data.iloc[:,-1].values # класс для PUL ( P & U)
y_positive = mod_data.iloc[:,-2].values # истиная разметка
```

```python
# random negative sampling

mod_data = mod_data.sample(frac=1)
neg_sample = mod_data[mod_data['class_test']==-1][:len(mod_data[mod_data['class_test']==1])]
sample_test = mod_data[mod_data['class_test']==-1][len(mod_data[mod_data['class_test']==1]):]
pos_sample = mod_data[mod_data['class_test']==1]
print(neg_sample.shape, pos_sample.shape)
sample_train = pd.concat([neg_sample, pos_sample]).sample(frac=1)
```

```python
sample_train.head()
```

```python
model = GradientBoostingClassifier()
sample_train.loc[sample_train['class_test']==-1,'class_test']=0
model.fit(sample_train.iloc[:,:-2].values, 
          sample_train.iloc[:,-1].values)
y_predict = model.predict(sample_test.iloc[:,:-2].values)
f1, recall, precision = evaluate_results(sample_test.iloc[:,-2].values, y_predict)
```

```python
models_results['model'].append('RNS_0.9')
models_results['f1'].append(f1)
models_results['recall'].append(recall)
models_results['precision'].append(precision)
```
